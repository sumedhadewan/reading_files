## Parsing CSV file

### Using python

```python
def parse_file(datafile):
    with open(datafile ,'r') as f:
        header = f.readline().split(',')
        count = 0
        for line in f:
#             print(line)
            if count == 10:
                break
            print(line)
            count += 1
```

```python
parse_file('beatles-diskography.csv')
```

```python
def parse_file_into_dict(datafile):
    data = []
    with open(datafile,'r') as f:
        header = f.readline().split(',')
        count = 0
        for line in f:
#             if count == 14:
#                 break
#             count += 1
            fields = line.split(',')
            entry ={}
            for index,value in enumerate(header):
                entry[header[index].strip()] = fields[index].strip()
            data.append(entry)
    return data
            
```

```python
parse_file_into_dict('beatles-diskography.csv')
```

### Using Csv module

```python
import csv

DATAFILE = 'beatles-diskography.csv'

def parse_csv(datafile):
    data= []
    with open(datafile,'r') as sd:
        r = csv.DictReader(sd)
        for line in r:
            print(line)
            print('\n')
#             data.append(line)
#     return data
```

```python
parse_csv(DATAFILE)
```

CSV module deals with CSV format in pretty complete way.
**
DICTReader class from CSV module - read all data from the file into Dict.
Assume 1st row is header that use for field name.
As it read row/line , it will create Dict for each row. Field Name is whatever found in the 1st Row and value in return will be associated on each row of the line. Automatically deal with values like 'Parlophone(NZ), Capitol(US)'
