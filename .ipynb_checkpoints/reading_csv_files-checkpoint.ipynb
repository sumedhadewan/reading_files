```python
def parse_file_2(datafile):
    data = []
    with open(datafile,'r') as f:
        count = 0
        header = f.readline().split(',')
        for line in f:
            if count == 5:
                break
            fields = line.split(',')
            entry = {}
            for index,value in enumerate(fields):
                entry[fields[index].strip()] = value.strip()
            data.append(entry)    
            count += 1
    return data
```

```python
parse_file_2(datafile)
```

## Using CSV module

### DictModule
Automatically assume 1st line from datafile as header.
and returns each line as dict.Result same as above

```python
import csv
import os
DATADIR = ''
DATAFILE = 'beatles-diskography.csv'
datafile = os.path.join(DATADIR, DATAFILE)
print(datafile)
```

```python
def parse_csv_1(datafile):
    data= []
    with open(datafile,'r') as f:
        count = 0
        r = csv.DictReader(f)
        for line in r:
            if count == 5 :
                break
#             print(line)
            data.append(line)
            count += 1
    return data
```

```python
parse_csv_1(datafile)
```

```python
import os
import csv
datafile = '745090.csv'
```

```python
def parse_csv_2(datafile):
    data = []
    with open(datafile) as f:
        name = f.readline()
        header = f.readline()
        reader = csv.reader(f)
        count = 0
        for line in reader:
            if count == 5:
                break
#             print(line)
            data.append(line)
            count += 1
    return data
        
```

# Write to CSV file

```python
filename = 'out_file_1.csv'
```

```python
def save_file(data,filename):
    with open(filename,'w') as f:
        w = csv.writer(f ,delimiter='|')
        for line in data:
            w.writerow(line)
```
