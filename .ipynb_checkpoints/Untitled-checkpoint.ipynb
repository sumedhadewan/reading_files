```python
def decode(days):
    years, remaining_days = divmod(days,365)
    months, remaining_days = divmod(remaining_days,12)
    weeks, remaining_days = divmod(remaining_days,7)
    return('{} years {} months {} weeks'.format(years,months,weeks))
```

```python
decode(452)
```
