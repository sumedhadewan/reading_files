# List

#### List Methods

append : add/append items  , takes one argument

```python
items = ['Js','Python','PHP']
```

```python
print(items)
```

```python
items.append(['Ruby','css'])
```

```python
print(items)
```

clear : clear the entire list

```python
items.clear()
```

```python
print(items)
```

copy : shallow copy of list

```python
items = ['JS','Python','PHP']
copied_list = items.copy()
```

```python
print(items)
print(copied_list)
```

count : count the number of items occurs in a list , shallow level

```python
items.count('PHP')
```

```python
items2 = ['JS','Python','PHP',['PHP','Ruby']]
```

```python
items.count('PHP')
```

extend : add multiple items to a list

```python
print(items)
```

```python
items.extend(('CSS','HTML'))
```

```python
print(items)
```

index : return index of an element

```python
print(items)
```

```python
items.index('CSS')
```

```python
items2
```

```python
items2.index('PHP')
```

insert : insert an item in a given position
Lists in python starts from 0

```python
items
```

```python
items.insert(1,'Ruby')
```
