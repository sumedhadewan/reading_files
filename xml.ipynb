# Screen scarping

Some Analysis tasks , its not easy to get the data we need. The data is there somewhere but not in form we simply download, In these situations , data is usually found in bits and pieces that need to wrangle together. Sometime all pieces are one single website , sometimes we need to gather them from multiple sources.
As data wrangler we need to be nibble and creative to gather the data we need.

```python
from bs4 import BeautifulSoup
import requests
```

```python
def options(soup , id):
    option_values = []
    carrier_list = soup.find(id = id)
    for option in carrier_list.find_all('option'):
        option_values.append(option['value'])
    return option_values
```

```python
def print_list(label , codes):
    print(label)
    for c in codes:
        print(c)
```

```python
def extract_data(soup,id):
    return soup.find(id = id )['value']
    
```

```python
def main():
    soup = BeautifulSoup(open('united_boston.html'))
    carrier_codes = options(soup , 'CarrierList')
#     print_list('Carriers',carrier_codes)
    
    airport_codes = options(soup , 'AirportList')
#     print_list('Airports',airport_codes)
    viewstate = extract_data(soup,'__VIEWSTATE')
    eventvalidation = extract_data(soup,'__EVENTVALIDATION')
    viewstategenerator = extract_data(soup,'__VIEWSTATEGENERATOR')
    print(viewstategenerator)
    print(eventvalidation)
    print(viewstate)
    r = requests.post('https://www.transtats.bts.gov/Data_Elements.aspx?Data=2',
                        data = {
                            'AirportList' : 'BWI',
                            'CarrierList' : '9E',
                            'Submit' : 'Submit',
                            "__EVENTTARGET":"",
                            "__EVENTARGUMENT":"",
                            "__VIEWSTATE":viewstate,
                            "__EVENTVALIDATION" : eventvalidation,
                            "__VIEWSTATEGENRATOR": viewstategenerator
                        }
                     )
    return r.text
```

```python
main()
```
