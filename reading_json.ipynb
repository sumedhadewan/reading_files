```python
def pretty_print(data, indent = 4):
    if type(data) == dict:
        print(json.dumps(data, indent = indent ,sort_keys = True))
    else:
        print(data)
```

```python
def query_site(url,api_key):
    params = {'api-key' : api_key}
    r=requests.get(url, params = params)
    if r.status_code == 200:
        return r.json()
    else:
#         print('error')
        r.raise_for_status()
```

```python
def get_popular(url,kind,days,api_key):
    if days not in [1,7,30]:
        print('Time period can be 1,7,30 days only')
        return False
    if kind not in ['viewed','share','emailed']:
        print('kind can only be one of viewed/share/emailed')
    url += '{0}/{1}.json'.format(kind,days)
    return query_site(url,api_key)
```
