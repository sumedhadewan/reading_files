```python
def decode(days):
    years, remaining_days = divmod(days,365)
    months, remaining_days = divmod(remaining_days,12)
    weeks, remaining_days = divmod(remaining_days,7)
    return(f'{years} years {months} months {weeks}weeks')
#     return('{} years {} months {} weeks'.format(years,months,weeks))
```
