# List

#### List Methods

clear : clear the entire list

```python
items.clear()
```

```python
print(items)
```

copy : shallow copy of list

```python
items = ['JS','Python','PHP']
copied_list = items.copy()
```

```python
print(items)
print(copied_list)
```

count : count the number of items occurs in a list , shallow level

```python
items.count('PHP')
```

```python
items2 = ['JS','Python','PHP',['PHP','Ruby']]
```

```python
items.count('PHP')
```

index : return index of an element

```python
print(items)
```

```python
items.index('CSS')
```

```python
items2
```

```python
items2.index('PHP')
```
